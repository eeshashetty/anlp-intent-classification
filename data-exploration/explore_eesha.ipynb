```python
# Original Approach - use CSV
import csv
with open(FILEPATH) as f:
    split_reader = csv.reader(f, delimiter='\t')
    split_data = list(split_reader)
```

```python
text = [line[2] for line in split_data[1:]]
labels = [list(map(int, line[1])) for line in split_data[1:]]
```

```python
print(labels[0], text[0])
```

```python
final_df = pd.DataFrame(columns=["Text", "Labels"])
final_df["Text"] = text
final_df["Labels"] = labels
final_df.head()
```
